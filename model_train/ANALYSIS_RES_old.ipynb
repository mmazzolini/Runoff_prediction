```python
import pandas as pd
import numpy as np
```

```python
res=pd.read_csv('a_res.csv',index_col='id_station')[:18]
res
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
idx_max=res['Area_m^2'].idxmax()
f, ax = plt.subplots(figsize=(10, 7))

sns.scatterplot(x="Area_m^2", y="real_time",data=res,hue='id_station',legend=False,palette='tab10',marker='x',s=150).set(xscale="log")
sns.scatterplot(x="Area_m^2", y="10_days",data=res,hue='id_station',legend=False,palette='tab10',marker='_',s=150).set(xscale="log")
sns.scatterplot(x="Area_m^2", y="20_days",data=res,hue='id_station',legend=False,palette='tab10',marker='1',s=150).set(xscale="log")
sns.scatterplot(x="Area_m^2", y="clim",data=res,hue='id_station',legend=False,palette='tab10',marker='o',s=50).set(xscale="log")

x_off=+10000000000/2
y_off=-0.009
a=0.8

plt.text((res['Area_m^2'].loc[idx_max]+x_off),(res['20_days'].loc[idx_max]+y_off),' 20 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['Area_m^2'].loc[idx_max]+x_off,res['10_days'].loc[idx_max]+y_off ,' 10 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['Area_m^2'].loc[idx_max]+x_off,res['real_time'].loc[idx_max]+y_off ,' real time', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['Area_m^2'].loc[idx_max]+x_off,res['clim'].loc[idx_max]+y_off ,' climatology', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.vlines(x=np.array(res['Area_m^2']), ymin=np.array(res['20_days']),ymax=np.array(res.real_time),colors=sns.color_palette(palette='tab10',n_colors=res.shape[0]),
          linestyle='dashed')
plt.ylabel('r^2 [/]')

f.legend=None
```

```python
idx_max=res['avg_h'].idxmax()

f, ax = plt.subplots(figsize=(10, 7))

sns.scatterplot(x="avg_h", y="real_time",data=res,hue='id_station',legend=False,palette='tab10',marker='x',s=150)
sns.scatterplot(x="avg_h", y="10_days",data=res,hue='id_station',legend=False,palette='tab10',marker='_',s=150)
sns.scatterplot(x="avg_h", y="20_days",data=res,hue='id_station',legend=False,palette='tab10',marker='1',s=150)
sns.scatterplot(x="avg_h", y="clim",data=res,hue='id_station',legend=False,palette='tab10',marker='o',s=50)

x_off=10
y_off=-0.015
a=0.8

plt.text((res['avg_h'].loc[idx_max]+x_off),(res['20_days'].loc[idx_max]+y_off),' 20 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['avg_h'].loc[idx_max]+x_off,res['10_days'].loc[idx_max]+y_off ,' 10 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['avg_h'].loc[idx_max]+x_off,res['real_time'].loc[idx_max]+y_off ,' real time', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['avg_h'].loc[idx_max]+x_off,res['clim'].loc[idx_max]+y_off ,' climatology', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.vlines(x=np.array(res['avg_h']), ymin=np.array(res['20_days']),ymax=np.array(res.real_time),colors=sns.color_palette(palette='tab10',n_colors=res.shape[0]),
          linestyle='dashed')

plt.ylabel('r^2 [/]')

f.legend=None
```
