```python
import pandas as pd
import numpy as np
```

```python
area=pd.read_csv('id_stat_area.csv',index_col='id_station')
area.index.name='STAT_CODE'
area.rename(columns={"SURFACE": "Area_m^2"},inplace=True)
area
```

```python
r=pd.read_csv('r_res.csv',index_col='STAT_CODE')
r.drop(columns='measured runoff',inplace=True)
r.rename(columns={"model output": "real_time", 
                  "output 1 lead time":"10_days", 
                  "output 2 lead time":"20_days",
                 "runoff climatology": "clim"},inplace=True)

r
```

```python
avgh=pd.read_csv('id_stat_avgh.csv',index_col='name')
avgh.index.name='STAT_CODE'
avgh.drop(columns='Unnamed: 0',inplace=True)
avgh
```

```python
res=r.merge(area, how='outer', right_index=True,left_index=True).merge(avgh,how='outer',right_index=True,left_index=True)
res
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
idx_max=res['Area_m^2'].idxmax()
f, ax = plt.subplots(figsize=(10, 7))

sns.scatterplot(x="Area_m^2", y="real_time",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='x',s=150).set(xscale="log")
sns.scatterplot(x="Area_m^2", y="10_days",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='_',s=150).set(xscale="log")
sns.scatterplot(x="Area_m^2", y="20_days",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='1',s=150).set(xscale="log")
sns.scatterplot(x="Area_m^2", y="clim",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='o',s=50).set(xscale="log")

x_off=+10000000000/2
y_off=-0.009
a=0.8

plt.text((res['Area_m^2'].loc[idx_max]+x_off),(res['20_days'].loc[idx_max]+y_off),' 20 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['Area_m^2'].loc[idx_max]+x_off,res['10_days'].loc[idx_max]+y_off ,' 10 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['Area_m^2'].loc[idx_max]+x_off,res['real_time'].loc[idx_max]+y_off ,' real time', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['Area_m^2'].loc[idx_max]+x_off,res['clim'].loc[idx_max]+y_off ,' climatology', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.vlines(x=np.array(res['Area_m^2']), ymin=np.array(res['20_days']),ymax=np.array(res.real_time),colors=sns.color_palette(palette='tab10',n_colors=res.shape[0]),
          linestyle='dashed')

plt.ylabel('r^2 [/]')

f.legend=None
```

```python
idx_max=res['mean_h'].idxmax()

f, ax = plt.subplots(figsize=(10, 7))

sns.scatterplot(x="mean_h", y="real_time",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='x',s=150)
sns.scatterplot(x="mean_h", y="10_days",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='_',s=150)
sns.scatterplot(x="mean_h", y="20_days",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='1',s=150)
sns.scatterplot(x="mean_h", y="clim",data=res,hue='STAT_CODE',legend=False,palette='tab10',marker='o',s=50)

x_off=10
y_off=-0.015
a=0.8

plt.text((res['mean_h'].loc[idx_max]+x_off),(res['20_days'].loc[idx_max]+y_off),' 20 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['mean_h'].loc[idx_max]+x_off,res['10_days'].loc[idx_max]+y_off ,' 10 days', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['mean_h'].loc[idx_max]+x_off,res['real_time'].loc[idx_max]+y_off ,' real time', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.text(res['mean_h'].loc[idx_max]+x_off,res['clim'].loc[idx_max]+y_off ,' climatology', 
 fontdict=dict(color='black', alpha=a, size=12))

plt.vlines(x=np.array(res['mean_h']), ymin=np.array(res['20_days']),ymax=np.array(res.real_time),colors=sns.color_palette(palette='tab10',n_colors=res.shape[0]),
          linestyle='dashed')

plt.ylabel('r^2 [/]')

f.legend=None
```
