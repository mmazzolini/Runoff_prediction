```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
path='results\SAN_BENIGNO__10.csv'
r=pd.read_csv(path)
#r_a=r_a.drop(columns='D) no snow-runoff')
```

```python
ax,fig=plt.subplots(figsize=(10,3.5))

r.iloc[2:].drop(columns=['t_unit']).plot.line(ax=plt.gca())

#plt.hlines(y=1,xmin=2,xmax=7,linestyles='dashed')
plt.hlines(y=r.iloc[1][2],xmin=2,xmax=7,linestyles='dashed',label='climatology')
plt.xticks(ticks = range(2,8) ,labels=list(np.arange(0,r.t_unit[0]*6,r.t_unit[0])))
plt.legend()
plt.ylabel('R^2 [/]')
plt.xlabel('lead time [days]')
plt.title('SVM (rbf) scores')
```
