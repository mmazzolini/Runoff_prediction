### merge multiple snowgrid yearly files

```python
from pathlib import Path
from netCDF4 import Dataset
import numpy as np
import xarray as xr
src_folder = 'Z:\ADO\ZAMG\SNOWGRID\\'
from os import walk
```

```python
from os import walk
mypath= src_folder

filenames = next(walk(mypath), (None, None, []))[2]  # [] if no file
```

```python
filenames
```

```python
c=0
for i in filenames:
    if c==0:
        ds = xr.open_dataset(mypath+i)
    else:
        ds_add = xr.open_dataset(mypath+i)
        ds=xr.concat([ds, ds_add], dim="time")
        
    c=c+1
```
